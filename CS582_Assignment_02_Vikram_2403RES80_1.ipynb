# What is Customer Churn?

**Customer churn**, also known as customer attrition or customer turnover, refers to the phenomenon where customers stop doing business with a company or discontinue using its products or services. It’s a critical metric for businesses, especially in subscription-based models or industries with recurring revenue, because it directly impacts revenue and growth.



Understanding customer churn is crucial for businesses to identify patterns, factors, and indicators that contribute to customer attrition. By analyzing churn behavior and its associated features, companies can develop strategies to retain existing customers, improve customer satisfaction, and reduce customer turnover. Predictive modeling techniques can also be applied to forecast and proactively address potential churn, enabling companies to take proactive measures to retain at-risk customers.

Customer-Churn.png

**Key Aspects of Customer Churn**

**1. Definition:**

---



  1.   **Churn Rate:** This is the percentage of customers who leave or stop using a product/service during a given time period. For instance, if a company has 1000 customers at the start of the year and 100 of them stop using the service by the end of the year, the churn rate is 10%.

    **Churn Rate = Number of Customers Lost / Total Number of Customers at the Start of the Period**

**2. ​Types of Churn:**

---



1.   **Voluntary Churn:** Customers who actively decide to leave. This might be due to dissatisfaction, better alternatives, or other reasons.
2.   **Involuntary Churn:** Customers who leave due to factors beyond their control, such as payment issues, or technical problems that prevent them from using the service.

**3. Implications for Businesses:**

---

1.   **Revenue Impact:** Losing customers means losing revenue. For subscription-based businesses, churn directly affects recurring revenue.
2. **Acquisition Costs:** Acquiring new customers often costs more than retaining existing ones. High churn rates can lead to increased marketing and sales expenses.
3. **Customer Lifetime Value (CLV):** A high churn rate reduces the average CLV, which is the total revenue a business can expect from a customer over their lifetime.

**4. Causes of Churn:**

---



1. **Poor Customer Service:** Lack of support or poor service experiences can drive customers away.
2. **Product Issues:** Problems with the product or service, such as bugs, lack of features, or poor usability, can contribute to churn.
3. **Competitive Offers:** Customers might leave for better or cheaper alternatives provided by competitors.
4. **Pricing:** Changes in pricing or perceived value can lead to churn if customers feel they are not getting their money’s worth.

**5. Managing and Reducing Churn:**

---



1. **Customer Feedback:** Regularly gather and analyze customer feedback to identify and address issues before they lead to churn.
2. **Personalization:** Tailor experiences and communications to meet the individual needs of customers.
3. **Loyalty Programs:** Implement loyalty programs or incentives to encourage customers to stay.
4. **Predictive Analytics:** Use data analysis to identify at-risk customers and take proactive measures to retain them.

**6. Predictive Modeling for Churn:**

---



1. **Customer Churn Prediction:** Companies often use machine learning and statistical models to predict which customers are likely to churn. By identifying these customers early, businesses can take targeted actions to retain them.

# **Models I have considered using is ANN**

**Building and Training ANN Models**

Create and train two different neural network models: one with a single hidden layer and one with two hidden layers.



**1. Model with One Hidden Layer**

Define a neural network with one hidden layer:

```python
# Model with one hidden layer
model_1_hidden = Sequential([
    # Input layer and first hidden layer
    Dense(64, input_dim=X_train.shape[1], activation='relu'),
    # Output layer with sigmoid activation for binary classification
    Dense(1, activation='sigmoid')
])

# Compile the model
model_1_hidden.compile(optimizer='adam',  # Optimizer
                       loss='binary_crossentropy',  # Loss function for binary classification
                       metrics=['accuracy'])  # Evaluation metric

history_1_hidden = model_1_hidden.fit(
    X_train,  # Training features
    y_train,  # Training target
    epochs=100,  # Number of epochs to train
    batch_size=32,  # Number of samples per gradient update
    validation_data=(X_val, y_val),  # Use validation data
    verbose=1  # Display training progress
)
```

**2. Model with Two Hidden Layers**

Define a neural network with two hidden layers:

```python
# Define the ANN model with two hidden layers
model_2_hidden = Sequential([
    # Input layer and first hidden layer
    Dense(64, input_dim=X_train.shape[1], activation='relu'),
    # Second hidden layer
    Dense(64, activation='relu'),
    # Output layer with sigmoid activation for binary classification
    Dense(1, activation='sigmoid')
])

# Compile the model
model_2_hidden.compile(optimizer='adam',  # Optimizer
                       loss='binary_crossentropy',  # Loss function for binary classification
                       metrics=['accuracy'])  # Evaluation metric

# Train the model
history_2_hidden = model_2_hidden.fit(
    X_train,  # Training features
    y_train,  # Training target
    epochs=50,  # Number of epochs to train
    batch_size=32,  # Number of samples per gradient update
    validation_split=0.2,  # Use 20% of training data for validation
    verbose=1  # Display training progress
)
```

# **Evaluate the Models**

Assess the performance of the models using the test set:

```python
# Evaluate the model with one hidden layer
loss_1, accuracy_1 = model_1_hidden.evaluate(X_test, y_test, verbose=1)
print(f"Model with 1 hidden layer - Test Loss: {loss_1:.4f}, Test Accuracy: {accuracy_1:.4f}")

# Evaluate the model with two hidden layers
loss_2, accuracy_2 = model_2_hidden.evaluate(X_test, y_test, verbose=1)
print(f"Model with 2 hidden layers - Test Loss: {loss_2:.4f}, Test Accuracy: {accuracy_2:.4f}")
```

# **Plot Training History**

Visualize the training and validation metrics to understand how well the models are learning:

```python
# Function to plot training history
def plot_history(history, title):
    plt.figure(figsize=(14, 6))

    # Plot accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'{title} - Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    # Plot loss
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'{title} - Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Plot training history for both models
plot_history(history_1_hidden, 'Model with 1 Hidden Layer')
plot_history(history_2_hidden, 'Model with 2 Hidden Layers')
```
